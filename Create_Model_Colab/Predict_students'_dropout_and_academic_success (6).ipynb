```python
import pandas as pd

###### # Re-loading the dataset using ';' or ';' as the delimiter
data = pd.read_csv("data.csv", delimiter=';')
```

```python
# Displaying the first few rows of the dataset
data.describe()
```

```python
# Filtramos el DataFrame para excluir las filas con "Enrolled" en la columna "Target"
data_test = data[data['Target'] != 'Enrolled']
data = data[data['Target'] != 'Enrolled']
len(data)
```

```python
# Checking for missing values in the dataset
missing_values = data.isnull().sum()

# Displaying columns with missing values (if any)
missing_values[missing_values > 0]
```

```python
from sklearn.preprocessing import StandardScaler

# Encoding categorical variables using one-hot encoding
data_encoded = pd.get_dummies(data, drop_first=True)

# Extracting features and target variables
X = data_encoded.drop(columns=['Target_Desertor','Target_Graduate'], errors='ignore')
y = data['Target'].map({'Dropout': 1, 'Graduate': 0})

# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Displaying the first few rows of the encoded dataset
data_encoded.head()
```

```python
# We do this to save these columns and use it in flask app
import joblib

# Guardar las columnas del dataframe codificado en una lista
training_columns = X.columns.tolist()

# Guardar la lista en un archivo .pkl
joblib.dump(training_columns, 'training_columns.pkl')
print(training_columns)
```

#Training the model

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

X_train.shape, X_test.shape
```

```python
# Checking for NaN values in the target variable 'y'
nan_rows = y[y.isnull()]

nan_rows
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression

# Training the logistic regression model on cleaned data
logistic_model = LogisticRegression(max_iter=10000)
logistic_model.fit(X_train, y_train)

# Predicting on the cleaned test set
y_pred = logistic_model.predict(X_test)

# Calculating accuracy for cleaned data
accuracy = accuracy_score(y_test, y_pred)

# Displaying classification report and accuracy for cleaned data
classification_rep= classification_report(y_test, y_pred)
confusion_mat = confusion_matrix(y_test, y_pred)

print("Accuracy: ",accuracy)
print(classification_rep)
print("Confusion Matrix: \n",confusion_mat)
```

```python
import joblib

# Save the model
joblib.dump(logistic_model, 'logistic_model.pkl')

# Save the scaler
joblib.dump(scaler, 'scaler.pkl')
```

# Make predictions

```python
import joblib

# Carga el modelo y el escaler
model = joblib.load('logistic_model.pkl')
scaler = joblib.load('scaler.pkl')
```

```python
# 1. Select some rows as an example (here I am selecting the first 30)
sample_data = data_test.iloc[:30].drop(columns=['Target'])  # assuming that 'Target' is your output column

# 2. Preprocessing
# Performs the one-hot encoding
sample_encoded = pd.get_dummies(sample_data)
sample_encoded = sample_encoded.drop(columns=['Target_Desertor','Target_Graduate'], errors='ignore')
# Make sure that all expected columns are present in the test dataset
for col in training_columns:
    if col not in sample_encoded.columns:
        sample_encoded[col] = 0

# Reorder the columns
sample_encoded = sample_encoded[training_columns]

# Escale features
sample_scaled = scaler.transform(sample_encoded)

# 3. Prediction
predictions = model.predict(sample_scaled)

# 4. Evaluate
print("Predicciones:", predictions)
print("Valores reales:\n", data_test['Target'].iloc[:30].values)
```
